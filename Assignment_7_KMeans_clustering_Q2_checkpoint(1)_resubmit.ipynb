Model Creation

```python
# Normalization of the data
def norm_df(i):
    x = (i-i.min())/(i.min()-i.max())
    return(x)
```

```python
normal_data=norm_df(data)
normal_data
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for k in range(1,11):
    clf=KMeans(n_clusters=k)
    clf.fit(normal_data)
    wcss.append(clf.inertia_)
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("WCSS")
plt.ylabel("No of Clusters")
plt.show()
```

Here We Assume That k-value is to be 5

```python



from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
model1= KMeans(n_clusters=4)
model1.fit(data)
pred=model1.labels_
data['kmeans_cluster'] = pred

# Perform Hierarchical Clustering
hierarchical_model = AgglomerativeClustering(n_clusters=3)
hierarchical_labels = hierarchical_model.fit_predict(data)
data['Hierarchical_Clusters'] = hierarchical_labels

# Perform DBSCAN Clustering
dbscan_model = DBSCAN(eps=1.5, min_samples=5)
dbscan_labels = dbscan_model.fit_predict(data)
data['DBSCAN_Clusters'] = dbscan_labels

data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame with clustering results
plt.figure(figsize=(18, 5))

# Select relevant features and cluster labels
data_features = data[['Balance','Qual_miles','Bonus_miles','Bonus_trans','Flight_miles_12mo','kmeans_cluster','Hierarchical_Clusters','DBSCAN_Clusters']]

# Create pair plots
sns.pairplot(data_features, hue='kmeans_cluster', palette='viridis', diag_kind='kde', vars=['Balance','Qual_miles','Bonus_miles','Bonus_trans','Flight_miles_12mo','kmeans_cluster','Hierarchical_Clusters','DBSCAN_Clusters'])
plt.suptitle('Pair Plot for KMeans Clustering', y=1.02)
plt.show()

# Hierarchical Clustering
# Create pair plots
sns.pairplot(data_features, hue='Hierarchical_Clusters', palette='viridis', diag_kind='kde', vars=['Balance','Qual_miles','Bonus_miles','Bonus_trans','Flight_miles_12mo','kmeans_cluster','Hierarchical_Clusters','DBSCAN_Clusters'])
plt.suptitle('Pair Plot for Hierarchical_Clusters Clustering', y=1.02)
plt.show()

# DBSCAN Clustering
# Create pair plots
sns.pairplot(data_features, hue='DBSCAN_Clusters', palette='viridis', diag_kind='kde', vars=['Balance','Qual_miles','Bonus_miles','Bonus_trans','Flight_miles_12mo','kmeans_cluster','Hierarchical_Clusters','DBSCAN_Clusters'])
plt.suptitle('Pair Plot for DBSCAN Clustering', y=1.02)
plt.show()
```
